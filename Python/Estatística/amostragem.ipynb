```python
import pandas as pd
import numpy as np
from math import ceil
from sklearn.model_selection import train_test_split
```

# Amostragem simples

```python
url = "https://raw.githubusercontent.com/anaiza-z/curso-ciencia-de-dados/main/Dados/iris.csv"
base = pd.read_csv(url)
base
```

```python
#obter os mesmos resultados posteriormente 
np.random.seed(2345)
#gerando 150 amostras, de 0 e 1, com reposição, probabilidades equivalentes
amostra = np.random.choice(a = [0, 1], size = 150, replace = True, p = [0.7, 0.3])
amostra
#len(amostra[amostra == 0]) 
#len(amostra[amostra == 1])
```

```python
base_final = base.loc[amostra == 0]
base_final
```

# Amostragem sistemática

```python
populacao = 150
amostra = 15
k = ceil(populacao/amostra)
print(k)
```

```python
#gerando um inteiro aleatório entre 1 e k (inclusivo)
r = np.random.randint(low = 1, high = k + 1, size = 1)
print(r)
```

```python
#vai até 150 (populacao) de 10 em 10 (k) com o inicio definido em r
acumulador = r[0]
sorteados = []
for i in range(amostra):
    sorteados.append(acumulador)
    acumulador += k
print(sorteados)
```

```python
base_final = base.loc[sorteados]
base_final
```

# Amostragem estratificada

```python
base['class'].value_counts()
```

```python
#base.iloc[:, 0:4]: busca comente os atributos previsores
#base.iloc[:, 4]: busca somente a classe (espécie da planta)
#test_size: seleciona 50% dos dados, porém como essa função retorna 4 valores e só precisa de 50% (2 valores), então coloca "_"
#stratify: retorna a amostra baseada na classe

x, _, y, _ = train_test_split(base.iloc[:, 0:4], base.iloc[:, 4], test_size = 0.5, stratify = base.iloc[:, 4])
y.value_counts()
```
