```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

Dados normais

```python
#gerando uma variável com distribuição normal (1000 elementos)
dados = norm.rvs(size = 100)
dados
```

```python
#histograma
plt.hist(dados, bins = 20)
plt.title('Dados')
```

```python
#verificar a distribuição
fig, ax = plt.subplots()
stats.probplot(dados, fit = True, plot = ax)
plt.show()
```

```python
# p > 0,05 indica que os dados podem estar normalmente distribuídos 
stats.shapiro(dados)
```

Dados não normais

```python
dados2 = skewnorm.rvs(4, size = 100)
dados2
```

```python
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados2, fit = True, plot = ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
